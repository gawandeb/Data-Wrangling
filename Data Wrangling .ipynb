# Dealing with Missing values NA's

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('D:/Anaconda/Placements.csv') 
```

```python
display(df)
```

```python
#check whether the data has no values(NaN) in some cells 
# we need to remove NaN from the dat frame 
```

```python
print (df.isnull())  # check if the data frame has Null values (NaN)
```

```python
# Drop NaN values
# All the rows containing NaN values will be dropped 
print(df.dropna()) 
```

```python
# Drop columns which have NaN values 
print (df.dropna(axis=1))
```

```python
print (df.dropna(axis=1,how="all")) # if we want to drop the column if all the cells in that column contain NaN
```

```python
# threshold value parameter lets you specify the minimum number of non null value for the row/column to be kept
print (df.dropna(thresh=15))
```

# Filling Null Values (NA cells)

```python
# fill null values with zero
print(df.fillna(0))
```

```python
# specify a forward-fill to propagate previous value forward
print(df.fillna(method="ffill"))
```

```python
# fill forward column wise
print(df.fillna(method="ffill",axis=1))
```

```python
# back-fill to propagate the Na values backwards
print(df.fillna(method="bfill"))
```

```python
display(df.fillna(method="bfill"))
```

# Cross Tabulation
###  A cross-tabulation is a two- (or more) dimensional table that records the number (frequency) of respondents that have the specific characteristics described in the cells of the table.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:/Anaconda/Placements.csv')
df
```

```python
# for crosstabulation first tabulate the data on the basis to two columns
x=pd.crosstab(df['degree_t'],df['degree_p'],margins=True)
```

```python
(x.head(5))
```

```python
# perConvert is a python function for percentage conversion
def perConvert(var): 
    return var/float(var[-1])
```

```python
x=pd.crosstab(df['degree_p'],df['degree_t'],margins=True).apply(perConvert,axis=1)
(x.head(5))
```

# Pivoting 

### Create a new derived table out of an existing table
### Specify an index or indices to pivot on
### Reshape data based on column values
### uses Unique values from index/columns

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('D:/Anaconda/GlobalFirePower.csv')
df.head()
```

```python
# specify a unique index
# as ISO3 is a column with all unoque values we choose it as our unique index
x=pd.pivot_table(df,index=['ISO3'])
x.head()
```

```python
# specify two column indices
x = pd.pivot_table(df,index=['ISO3','Country'])
x.head()
```

```python
# include the corresponding values for analysis
x=pd.pivot_table(df,index=['Country','ISO3'], values=['Attack Aircraft','Combat Tanks'])
x.head(6)
```

```python
# more indices and values
x=pd.pivot_table(df,index=['Country','ISO3','Rank'], values=['Attack Helicopters','Submarines','Combat Tanks'])
x.head()
```

```python
# more indices and values
x=pd.pivot_table(df,index=['Country','ISO3','Rank'], values=['Attack Helicopters','Submarines','Combat Tanks'])
x.tail(8)
```

```python
## Pivot with indices that have non unique values
df=pd.read_csv('D:/Anaconda/Placements.csv')
df
```

```python
# pivot with indices that have non unique values
# sum the hsc_s according to the degree_t
x=pd.pivot_table(df,index=['hsc_s','degree_t'], aggfunc=np.sum)
x.head()
```

```python
x=pd.pivot_table(df,index=['hsc_s','degree_t'],values=["mba_p"], aggfunc=np.sum)
x.head()
```

# MERGE

```python
import pandas as pd
```

```python
gfp=pd.read_csv('D:/Anaconda/GlobalFirePower.csv')
gfp
```

```python
gdp=pd.read_csv('D:/Anaconda/gdp_csv.csv')
gdp
```

```python
gdp.columns=['Country','ISO3','Year','Value']
gdp
```

```python
# merge on columns that have common entries
#'default inner join'
x=pd.merge(gfp,gdp, on='Country')
```

```python
x.head(9)
```

```python
# Data base type merging 
```

```python
#keep every row in the left dataframe. where there are missing values of the 'on' VARIABLE on the right data frame, add empty
# LEFT JOIN
y = pd.merge(gdp,gfp[['Country','Manpower Available','Railway Coverage (km)']],
            on='Country', how='left')
```

```python
y.shape
```

```python
y.head()
```

```python
#keep every row in the Right dataframe. where there are missing values of the 'on' VARIABLE on the left data frame, add empty
# RIGHT JOIN
y = pd.merge(gdp,gfp[['Country','Manpower Available','Railway Coverage (km)']],
            on='Country', how='right')
```

```python
y.shape
```

```python
y.head()
```

```python
# A FULL OUTER JOIN returns all the rows from the left dataframe
# all the rows from the right dataframe, nad matched up rows where possible, with NaN's elsewhere
# indicator will tell us how the joining was done
#keep every row in the left dataframe. where there are missing values of the 'on' VARIABLE on the right data frame, add empty
# LEFT JOIN
y = pd.merge(gdp,gfp[['Country','Manpower Available','Railway Coverage (km)']],
            on='Country', how='outer', indicator = True)
```
